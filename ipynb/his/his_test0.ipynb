```python
his_20221022 = ts.TimeSeries("../../src/SWA/HIS/solo_L3_swa-his-comp-10min_20221022_V01.cdf")
his_20221020 = ts.TimeSeries("../../src/SWA/HIS/solo_L3_swa-his-comp-10min_20221020_V01.cdf")
his_20221021 = ts.TimeSeries("../../src/SWA/HIS/solo_L3_swa-his-comp-10min_20221021_V01.cdf")
his_20221023 = ts.TimeSeries("../../src/SWA/HIS/solo_L3_swa-his-comp-10min_20221023_V01.cdf")
his_20221024 = ts.TimeSeries("../../src/SWA/HIS/solo_L3_swa-his-comp-10min_20221024_V01.cdf")
his_20221025 = ts.TimeSeries("../../src/SWA/HIS/solo_L3_swa-his-comp-10min_20221025_V01.cdf")
his_20221026 = ts.TimeSeries("../../src/SWA/HIS/solo_L3_swa-his-comp-10min_20221026_V01.cdf")
```

```python
pas_ll_20221022 = ts.TimeSeries("../../src/SWA/PAS/solo_LL02_swa-pas-mom_20221022T040807-20221023T000035_V01I.cdf")
pas_ll_20221020 = ts.TimeSeries("../../src/SWA/PAS/solo_LL02_swa-pas-mom_20221020T040807-20221021T000039_V01I.cdf")
pas_ll_20221021 = ts.TimeSeries("../../src/SWA/PAS/solo_LL02_swa-pas-mom_20221021T000215-20221021T235219_V01C.cdf")
pas_ll_20221023 = ts.TimeSeries("../../src/SWA/PAS/solo_LL02_swa-pas-mom_20221023T000211-20221023T235355_V01C.cdf")
pas_ll_20221024 = ts.TimeSeries("../../src/SWA/PAS/solo_LL02_swa-pas-mom_20221024T040819-20221025T000036_V01I.cdf")
pas_ll_20221025 = ts.TimeSeries("../../src/SWA/PAS/solo_LL02_swa-pas-mom_20221025T000212-20221025T235216_V01I.cdf")
pas_ll_20221026 = ts.TimeSeries("../../src/SWA/PAS/solo_LL02_swa-pas-mom_20221026T040808-20221026T235216_V01I.cdf")
```

```python
pas_ll_20221022.to_dataframe()
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize=(8,6), layout="constrained", sharex=True)
for his_dataset in [his_20221020, his_20221021, his_20221022, his_20221023, his_20221024, his_20221025, his_20221026]:
    ax1.plot(his_dataset.to_dataframe().index.values, his_dataset.to_dataframe()["FE_O_ABUN"].values/10**(7.46 - 8.69),
             color="red", lw=0.5)
    ax1.fill_between(his_dataset.to_dataframe().index.values,
                      (his_dataset.to_dataframe()["FE_O_ABUN"].values - his_dataset.to_dataframe()["FE_O_ABUN_ERR"].values)/10**(7.46 - 8.69),
                      (his_dataset.to_dataframe()["FE_O_ABUN"].values + his_dataset.to_dataframe()["FE_O_ABUN_ERR"].values)/10**(7.46 - 8.69),
                      color="red", alpha=0.5)
    
    ax2.plot(his_dataset.to_dataframe().index.values, his_dataset.to_dataframe()["O7_O6_RATIO"].values,
                color="orange", lw=0.5)
    ax2.fill_between(his_dataset.to_dataframe().index.values,
                        (his_dataset.to_dataframe()["O7_O6_RATIO"].values - his_dataset.to_dataframe()["O7_O6_RATIO_ERR"].values),
                        (his_dataset.to_dataframe()["O7_O6_RATIO"].values + his_dataset.to_dataframe()["O7_O6_RATIO_ERR"].values),
                        color="orange", alpha=0.5)


for pas_dataset in (pas_ll_20221020, pas_ll_20221021, pas_ll_20221022, pas_ll_20221023, pas_ll_20221024, pas_ll_20221025, pas_ll_20221026):
    ax3.plot(pas_dataset.to_dataframe().index.values, np.sqrt(pas_dataset.to_dataframe()["SWA_PAS_VELOCITY_RTN_0"].values**2 + \
                                                              pas_dataset.to_dataframe()["SWA_PAS_VELOCITY_RTN_1"].values**2 + \
                                                                pas_dataset.to_dataframe()["SWA_PAS_VELOCITY_RTN_2"].values**2),
                                                                color="blue", lw=0.5)
    
    ax4.plot(pas_dataset.to_dataframe().index.values, pas_dataset.to_dataframe()["SWA_PAS_DENSITY"].values,
             color="green", lw=0.5)
    
for ax_ in (ax1,ax2,ax3):
    ax_.tick_params(axis="x", labelbottom=False)
ax4.tick_params(axis="x", rotation=45)

ax1.set_ylabel(r"$\rm [Fe/O]/[Fe/O]_{phot}$")
ax2.set_ylabel(r"$\rm O^{7+}/O^{6+}$")
ax3.set_ylabel(r"$v$ (km/s)")
ax4.set_ylabel(r"$n$ (cm$^{-3}$)")

```
