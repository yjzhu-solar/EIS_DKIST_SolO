```python
import numpy as np
import matplotlib.pyplot as plt
import astropy
from astropy.io import fits
from astropy.visualization import (AsinhStretch, ImageNormalize)
import sunpy
import sunpy.map
```

```python
iris_raster_file = "../../src/IRIS/20221024/2322/iris_l2_20221024_232249_3600609177_raster_t000_r00000.fits"
iris_sji1330_file = "../../src/IRIS/20221024/2322/iris_l2_20221024_232249_3600609177_SJI_1330_t000.fits"
iris_sji1400_file = "../../src/IRIS/20221024/2322/iris_l2_20221024_232249_3600609177_SJI_1400_t000.fits"
iris_sji2796_file = "../../src/IRIS/20221024/2322/iris_l2_20221024_232249_3600609177_SJI_2796_t000.fits"
```

```python
with fits.open(iris_raster_file) as hdul:
    iris_cii_spec = hdul[1].data[100,:,10:-10].copy()
    iris_siiv_spec = hdul[3].data[100,:,10:-10].copy()
    iris_mgii_spec = hdul[6].data[100,:,10:-10].copy()
```

```python
with fits.open(iris_sji1330_file) as hdul:
    hdul.info()
    iris_sji1330_img = hdul[0].data[100,:,:].copy()

with fits.open(iris_sji1400_file) as hdul:
    iris_sji1400_img = hdul[0].data[100,:,:].copy()

with fits.open(iris_sji2796_file) as hdul:
    iris_sji2796_img = hdul[0].data[100,:,:].copy()
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(8,5))

ax1.imshow(iris_cii_spec[70:90,:], aspect="auto",origin='lower',norm=ImageNormalize(vmin=0,vmax=1e2,stretch=AsinhStretch()),
           cmap="irissji1330")

ax2.imshow(iris_sji1330_img[70:90,300:380], aspect="auto",origin='lower',norm=ImageNormalize(vmin=0,vmax=1e2,stretch=AsinhStretch()),
              cmap="irissji1330")

plt.subplots_adjust(wspace=0, hspace=0)
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(8,5))

ax1.imshow(iris_siiv_spec[115:125,:], aspect="auto",origin='lower',norm=ImageNormalize(vmin=10,vmax=1e2),
              cmap="irissji1400")

ax2.imshow(iris_sji1400_img[115:125,300:380], aspect="auto",origin='lower',norm=ImageNormalize(vmin=10,vmax=1e2),
                cmap="irissji1400")

ax3.imshow(iris_siiv_spec[382:392,:], aspect="auto",origin='lower',norm=ImageNormalize(vmin=10,vmax=1e2),
              cmap="irissji1400")

ax4.imshow(iris_sji1400_img[382:392,300:380], aspect="auto",origin='lower',norm=ImageNormalize(vmin=10,vmax=1e2),
                cmap="irissji1400")

plt.subplots_adjust(wspace=0, hspace=0)
```

```python
plt.plot(np.nanmean(iris_siiv_spec[:,:],axis=1))
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(5,5))

ax1.imshow(iris_mgii_spec[115:125,:], aspect="auto",origin='lower',norm=ImageNormalize(vmin=100,vmax=1e3),
                cmap="irissjiNUV")

ax2.imshow(iris_sji2796_img[115:125,300:380], aspect="auto",origin='lower',norm=ImageNormalize(vmin=100,vmax=5e2),
                cmap="irissji2796")

ax3.imshow(iris_mgii_spec[382:392,:], aspect="auto",origin='lower',norm=ImageNormalize(vmin=100,vmax=1e3),
                cmap="irissjiNUV")

ax4.imshow(iris_sji2796_img[382:392,300:380], aspect="auto",origin='lower',norm=ImageNormalize(vmin=100,vmax=5e2),
                cmap="irissji2796")

plt.subplots_adjust(wspace=0, hspace=0)

for ax_ in (ax1,ax2,ax3,ax4):
    ax_.tick_params(axis='x', which='both',
                    labelbottom=False, labeltop=False, labelleft=False, labelright=False,direction="in")
    ax_.tick_params(axis='y', which='both',
                    labelbottom=False, labeltop=False, labelleft=False, labelright=False,direction="in")
    ax_.grid("on")
```
