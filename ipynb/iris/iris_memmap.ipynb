```python
import irispy
import irispy.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
irispy.__version__
```

```python
irispy_cube = irispy.io.read_files("../../src/IRIS/20221024/2322/iris_l2_20221024_232249_3600609177_SJI_1330_t000.fits")
```

```python
irispy_cube_memmap = irispy.io.read_files("../../src/IRIS/20221024/2322/iris_l2_20221024_232249_3600609177_SJI_1330_t000.fits", memmap=True)
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(10,5),layout='constrained')
_ = ax1.hist(irispy_cube.data.flatten(), bins=100, histtype='step', color='k',range=(0,500))
_ = ax2.hist(irispy_cube_memmap.data.flatten(), bins=100, histtype='step', color='k')
ax2.axvline(-32768, color='r', linestyle='--')

for ax_ in (ax1,ax2):
    ax_.set_yscale('log')

ax1.set_title("memmap=False")
ax2.set_title("memmap=True")
```
